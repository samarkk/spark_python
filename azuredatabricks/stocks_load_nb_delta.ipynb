```
# get the futures and options processed table
# do a show to verify data exists
fopr_reg = spark.read.table('fopr')
fopr_reg.show()
```

```
# create a blob container foprdelta
# mount it in azure databricks using mount name /mnt/foprdelta-azkv - see the notebook mount_adls2_storage
# select year 2018 and month = 1 and write it to the mount point above as an external table
```

```
# for exploration filter fopr_reg to year 2018, month 1
fopr_reg.filter("year = 2018 and month = 1").show()
```

```
# just checking the count
fopr_reg.filter("year = 2018 and month = 1").count()
```

```
# save the filtered fo processed table as an external delta table in the mounted location
fopr_reg.filter("year = 2018 and month = 1").write.format('delta').option('path', '/mnt/foprdelta-azkv').saveAsTable('foprdel')
```

```
# verify that the adls2 container is showing data for the external table created above
```

```
%sql
-- read from the delta table
select * from foprdel limit 5
```

```
# read from the directory location directly
spark.read.format('delta').load('/mnt/foprdelta-azkv').show()
```

```
# save jan to mar 2018 processed data as a partitioned delta table
fopr_reg.filter("year = 2018 and month between 1 and 3").write.format('delta')\
.option('path', '/mnt/foprdeltapart-azkv')\
.partitionBy('year', 'month', 'day').saveAsTable('foprdelpart')
```

```
# verify in the target directory partioned structure for the delta table is present
```

```
%sql
select * from foprdel where year = 2018 and month = 1 and symbol = 'ACC'
```

```
%sql
-- update the foprdel table
update foprdel 
set symbol = 'ACCEMENT'
where year = 2018 and month = 1 and symbol = 'ACC'
```

```
%sql
-- verify that the symbol ACC is updated to ACCEMENT
select * from foprdel where symbol = 'ACCEMENT'
```

```
%sql
-- Delete the updated rows from the foprdel table
Delete from foprdel where symbol = 'ACCEMENT'
```

```
%sql 
-- insert back ACC rows into foprdel table
insert into foprdel 
select * from fopr where year = 2018 and month = 1 and symbol = 'ACC'
```

```
%sql
-- as changes are made versions are saved
-- check the history of the delta table
DESC HISTORY foprdel
```

```
%sql
-- time travel to any version adding on the cluase VERSION AS OF
SELECT * FROM foprdel VERSION AS OF 1
WHERE SYMBOL LIKE 'ACC%'
```

```
%sql
-- time travel using timestamp
SELECT * FROM FOPRDEL TIMESTAMP AS OF '2022-01-19T10:43:16.000+0000'
where symbol like 'ACC%'
```

```
# python code to carry out time travel
spark.read.format('delta').option('timestampAsOf', '2022-01-19T10:43:16.000+0000').load('/mnt/foprdelta-azkv').show()
```

```
# time travel verifying ACC as restored is there
spark.read.format('delta')\
.option('timestampAsOf', '2022-01-19T10:43:16.000+0000')\
.load('/mnt/foprdelta-azkv')\
.filter("symbol like 'ACC%'")\
.show()
```

```
%sql
-- there may be GDBR or California Data Protection Requirements
-- we can vaccuum the table - which by default works with a retention period of 168 hours
-- force immediate vacuuming by setting retain time 0 hours and adding on the configuration shown below
VACUUM foprdel RETAIN 0 HOURS
```
